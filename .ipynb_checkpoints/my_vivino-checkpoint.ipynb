```python
# DATASET PREPROCESSING

df.head()
```

```python
# In this dataset quality range is between 3 and 8
```

We will divide quality range into two parts:

High quality wine: 6 - 8
Low quality wine: 3 - 5

```python
df["quality"] = np.where(df["quality"] > 5, 1, 0)
df["quality"]
```

```python
df.head()
```

```python
# we select dependent variable (label)
y = df["quality"]

# we select independent variable
x = df.drop("quality", axis = 1)
```

```python
! pip install Scikit-learn
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.25,
                                                    shuffle = True,
                                                    random_state = 1)
```

```python
norm = MinMaxScaler(feature_range = (0, 1))
norm.fit(x_train)
x_train = norm.transform(x_train)
x_test = norm.transform(x_test)
```

We will not use 'StandardScaler', because our dataset is not normally distributed. We will use 'MinMaxScaler' for normalizing this dataset. It transforms features by scaling each feature to a given range. This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between 0 and 1.

```python
print(x_train[0:5])
print("\n")
print(x_train.shape)
print("_________________________________________________________________________")
print(x_test[0:5])
print("\n")
print(x_test.shape)
```

```python
print(y_train.head())
print(y_train.shape)
print("___________________________")
print(y_test.head())
print(y_test.shape)
```

```python
# BUILDING CLASSIFICATION MODELS

# k-nearest neighbors algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()

knn_params = {"n_neighbors": np.arange(2, 50),
             "weights": ["uniform", "distance"],
             "leaf_size": [25, 30, 25]}

knn_cv_model = GridSearchCV(knn, knn_params, cv = 10)
knn_cv_model.fit(x_train, y_train)
```

```python
print("Best score for train set: " + str(knn_cv_model.best_score_))

print("____________________________________________")
```
